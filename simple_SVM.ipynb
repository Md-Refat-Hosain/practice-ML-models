```python
# Read the dataset
```

```python
df=pd.read_csv('/content/drive/MyDrive/Udemy ds course datasets/loan_small.csv')
```

```python
# Find out records with missing values
```

```python
df.isna().sum()
```

```python
#Drop missing values
```

```python
df = df.dropna()
```

```python
df
```

```python
# Drop irrelevant column
```

```python
df = df.drop(['Gender','Loan_ID'] , axis = 1)
```

```python
# create dummy variables
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df , drop_first= True)
```

```python
df
```

```python
# change bool to numeric
```

```python
df[['Area_semi','Area_urban','Loan_Status_Y']] = df[['Area_semi','Area_urban','Loan_Status_Y']].astype(int)
```

```python
df
```

```python
# divide dataframe for to X and y
```

```python
X = df.drop(['Loan_Status_Y'] , axis=1)
```

```python
y = df['Loan_Status_Y']
```

```python
# divide dataset for train and test
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= .3 , random_state= 42)
```

```python
# IMPORT and Build the SV classifier
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train,y_train)
```

```python
# Predict the outcome
```

```python
y_preds = svc.predict(X_test)
```

```python
# To compare the predictions Build confusion matrics
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test , y_preds)
```

```python
score = svc.score(X_test , y_test)
```

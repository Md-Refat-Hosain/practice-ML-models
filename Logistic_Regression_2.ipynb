```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
import statistics as st


from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

```python
# Reading data
```

```python
df = pd.read_csv('/content/drive/MyDrive/ud ds+ml (logistic regress....)/01Exercise1.csv')
```

```python
df
```

```python
LoanData = df.copy()
```

```python
# Identify the missing values and drop them
```

```python
LoanData.isna().sum()
```

```python
LoanData = LoanData.dropna()
```

```python
LoanData
```

```python
# Drop the irrelevent columns and genders
```

```python
LoanData = LoanData.drop(['gender'] , axis = 1)
```

```python
# Create Dummie variables
```

```python
LoanData = pd.get_dummies(LoanData , drop_first= True)
```

```python
LoanData
```

```python
# Changing boolean to int
```

```python
LoanData[['married_Yes','status_Y']] = LoanData[['married_Yes' , 'status_Y']].astype(int)
```

```python
LoanData
```

```python
# Normalize the data for Loan amount and income
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_ = StandardScaler()
```

```python
LoanData['income'] = scaler_.fit_transform(LoanData[['income']])
LoanData['loanamt'] = scaler_.fit_transform(LoanData[['loanamt']])
```

```python
LoanData
```

```python
# dividing the data into X and y
```

```python
X = LoanData.drop('status_Y' , axis = 1)
y = LoanData['status_Y']
```

```python
# spliting  X and y to train the model
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=.3 , random_state=42 , stratify=y)
```

```python
# Build the Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
y_preds = lr.predict(X_test)
```

```python
# Import confusion matrics
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test , y_preds)
cm
```

```python
# Score them
```

```python
score = lr.score(X_test , y_test)
```
